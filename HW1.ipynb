```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#
stats = []

#10
images = np.random.randint(0, 256, size=(10,64, 64, 3), dtype=np.uint8)#1064*64*3
gray_images = np.mean(images, axis=3).astype(np.uint8)

#10
plt.figure(figsize=(24,10))
for n in range(10):
    img = images[n]
    plt.subplot(2,5,n+1)
    plt.imshow(img, cmap='gray')
    plt.title("images"+ str(n+1))
    plt.axis("off")
plt.show()

#
plt.figure(figsize=(24,10))
for n in range(10):
    img = gray_images[n]
    max = np.max(img)             # 
    min = np.min(img)             # 
    mean = round(np.mean(img), 2) # round() 2 
    std = round(np.std(img), 2)   #  2 

    stats.append({
        'Max': max,
        'Min': min,
        'Mean': mean,
        'Std': std
    })
    plt.subplot(2,5,n+1)
    plt.imshow(img, cmap='gray')
    plt.title("gray_images"+ str(n+1),fontsize=10)
    plt.axis("off")

    
plt.show()

df = pd.DataFrame(stats)
df.to_excel(r"C:\Users\AN515\Desktop\NHW\HW1.xlsx")
```
