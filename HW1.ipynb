```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#開空矩陣
stats = []

#生10張
images = np.random.randint(0, 256, size=(10,64, 64, 3), dtype=np.uint8)#10張64*64*3
gray_images = np.mean(images, axis=3).astype(np.uint8)

#印出10張原圖
plt.figure(figsize=(24,10))
for n in range(10):
    img = images[n]
    plt.subplot(2,5,n+1)
    plt.imshow(img, cmap='gray')
    plt.title("images"+ str(n+1))
    plt.axis("off")
plt.show()

#取個數值、印出灰階
plt.figure(figsize=(24,10))
for n in range(10):
    img = gray_images[n]
    max = np.max(img)             # 取得影像中所有像素的最大值
    min = np.min(img)             # 取得影像中所有像素的最小值
    mean = round(np.mean(img), 2) # 計算影像所有像素的平均值，round()四捨五入到小數點後 2 位
    std = round(np.std(img), 2)   # 計算影像所有像素的標準差，並四捨五入到小數點後 2 位

    stats.append({
        'Max': max,
        'Min': min,
        'Mean': mean,
        'Std': std
    })
    plt.subplot(2,5,n+1)
    plt.imshow(img, cmap='gray')
    plt.title("gray_images"+ str(n+1),fontsize=10)
    plt.axis("off")

    
plt.show()

df = pd.DataFrame(stats)
df.to_excel(r"C:\Users\AN515\Desktop\NHW\HW1.xlsx")
```
