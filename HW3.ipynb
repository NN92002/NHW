```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

ogimg=cv2.imread(r"C:\Users\AN515\Desktop\NHW\input\cropHSV_yellow.jpg",-1) #bgr
#rgbimg=cv2.cvtColor(ogimg, cv2.COLOR_BGR2RGB)#rgb
hsvimg =cv2.cvtColor(ogimg, cv2.COLOR_RGB2HSV)#hsv
plt.imshow(ogimg)
plt.show()
```

```python
#濾波
med = cv2.medianBlur(hsvimg, 3)
# mean =cv2.blur(hsvimg, (3,3))
# gau =cv2.GaussianBlur(med3, (3,3),0)
```

```python
#遮罩
lowerb = np.array([30, 43, 46])
upperb = np.array([34, 255, 255])
# maskmean = cv2.inRange(mean, lowerb, upperb)
# maskgau = cv2.inRange(gau, lowerb, upperb)
maskmed = cv2.inRange(med, lowerb, upperb)

plt.imshow(maskmed,cmap='gray')
plt.show()
```

```python
#閉運算
kernel = np.ones((5, 5), np.uint8)
dilated_img = cv2.dilate(maskmed, kernel, iterations=1)#膨脹
eroded_img = cv2.erode(dilated_img, kernel, iterations=1)#侵蝕

plt.subplot(1,2,1)
plt.imshow(dilated_img, cmap='gray')
plt.title("dilated1")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(eroded_img, cmap='gray')
plt.title("eroded2")
plt.axis("off")
plt.show()
```
